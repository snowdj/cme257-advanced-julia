```julia
using Equations
using Interact
using Gadfly
```

Define expressions using Julia Symbols

```julia
Exp=(:x*:y + :x)^2
print(Exp)
```

Simplify expressions using simplify

```julia
print(simplify(Exp))
```

Can differentiate expressions, using the already defined "Der" relation

```julia
println(Der(:x*:x, :x )&relations["Der"])
println(Der( sqrt(:x), :x )&relations["Der"])
```

Can define equations using @equ macro and infer results using them

```julia
Eq1 = @equ c = a*b - 1
Eq2 = @equ f = a*c + 1/c
Eq3 = @equ a=2
print(Eq1&Eq2&Eq3)
```

Here is an example where interact.jl is used to generate plots of higher derivatives of some function

```julia
@manipulate for i = 0:5
    expr = sqrt(:x);
    for j = 1:i
        expr=Der(expr, :x )&relations["Der"]
    end
    plot( x -> evaluate(expr,Dict(:x=>x)),0, 2)
end
```
